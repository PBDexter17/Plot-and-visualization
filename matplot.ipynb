```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#正弦图
#x包含50个元素，均匀分布在{0，2pi}的区间上
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
plt.plot(x,y)
plt.show
```

```python
#一张纸里画多个图
plt.plot(x, y)
plt.plot(x, y*2)
plt.show()
```

```python
#调整样式
plt.plot(x, y, 'y*-')
plt.plot(x, y*2, 'm--')
plt.show()
```

```python
#控制图形大小
plt.figure(figsize=(6,3))
plt.plot(x, y)
plt.plot(x, y*2)
plt.show()
```

```python
#设置标题
plt.plot(x, y)
plt.plot(x, y*2)
plt.title('sin(x) & 2sin(x)')
plt.show()
```

```python
#设置坐标轴
plt.plot(x, y)
plt.plot(x, y*2)
plt.xlim((0, np.pi + 1))
plt.ylim((-3, 3))
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
#坐标轴刻度
plt.plot(x, y)
plt.plot(x, y*2)
plt.xticks((0, np.pi*0.5, np.pi, np.pi*1.5, np.pi*2))
plt.show()
```

```python
plt.plot(x, y, label='sin(x)')
plt.plot(x, y*2, label='2sin(x)')
plt.legend(loc='best')
plt.show()
```

```python
#添加注释
plt.plot(x, y)
x0 = np.pi
y0 = 0
plt.scatter(x0, y0, s=50)
plt.annotate('sin(np.pi)=%s' % y0, xy=(np.pi, 0), xycoords='data', xytext=(+30, -30), textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=.2'))
plt.text(0.5, -0.25, 'sin(np.pi)=0', fontdict={'size':16, 'color': 'r'})
plt.show()
```

```python
#使用子图
ax1 = plt.subplot(2, 2, 1) #行，列，活跃区
plt.plot(x, np.sin(x), 'r')
ax2 = plt.subplot(2, 2, 2, sharey=ax1) #与ax1共享y轴
plt.plot(x, 2*np.sin(x), 'g')
ax3 = plt.subplot(2, 2, 3)
plt.plot(x, np.cos(x), 'b')
ax4 = plt.subplot(2, 2, 4, sharey=ax3)
plt.plot(x, 2* np.cos(x), 'y')
plt.show()
```

```python
#调整子图大小
ax1 = plt.subplot(2, 1, 1) #行，列，活跃区
plt.plot(x, np.sin(x), 'r')
ax2 = plt.subplot(2, 3, 4, sharey=ax1) #与ax1共享y轴
plt.plot(x, 2*np.sin(x), 'g')
ax3 = plt.subplot(2, 3, 5)
plt.plot(x, np.cos(x), 'b')
ax4 = plt.subplot(2, 3, 6, sharey=ax3)
plt.plot(x, 2* np.cos(x), 'y')
plt.show()
```

```python
#散点图
k = 500
x = np.random.rand(k)
y = np.random.rand(k)
size = np.random.rand(k)*30 #每个点的大小
colour = np.arctan2(x, y) #每个点的颜色大小
plt.scatter(x, y, s=size, c=colour)
plt.colorbar() #添加颜色栏
plt.show()
```

```python
#柱状图
k = 10
x = np.arange(k)
y = np.random.rand(k)
plt.bar(x, y, color='blue')

for x, y in zip(x,y):
    plt.text(x, y, '%.2f' % y, ha='center', va='bottom') #标注数值
plt.show()
```
